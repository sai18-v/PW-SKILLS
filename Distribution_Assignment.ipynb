#4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

      Standardization (or Z-score) is a method of rescaling data so that it has:

        Mean = 0

        Standard deviation = 1

      It shows how many standard deviations a data point is away from the mean.

      Formula

            z=(x-μ)/σ

          Where:
            x = original value
            μ = mean
            σ = standard deviation

      Makes variables with different scales comparable

      Helps identify outliers

      Improves performance of many machine learning models

      Removes scale bias in data analysis

      

#5.What is Kurtosis and their type?

     Kurtosis is a statistical measure that describes the tailedness or peakedness of a data distribution. It indicates how extreme the values (outliers) are in a dataset.

     Types of Kurtosis:

      Mesokurtic

        Normal distribution

        Moderate peak and tails

     Leptokurtic

        High peak and heavy tails

        More extreme values (outliers)

      Platykurtic

        Flat peak and light tails

        Fewer extreme values

#6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die

      A uniform distribution is a good model for rolling a fair die because each outcome (1 to 6) has an equal probability of occurring.

      Since:

            Every face has the same chance (1/6)

            No outcome is favored

            Each roll is independent

      The outcomes are evenly distributed, which matches the definition of a uniform distribution.

#8. How does log transformation help in stabilizing variance and making data more normally distributed?

      Log transformation helps in data analysis by:

    1. Stabilizing Variance

      When data has large differences in scale, the variance increases with the mean (heteroscedasticity).
      Applying log reduces large values more than small values, making the spread more consistent.

    2. Making Data More Normal

      Right-skewed data (many small values, few large values) becomes more balanced after log transformation.
      This reduces skewness and makes the distribution closer to a normal distribution.
